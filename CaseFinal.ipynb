Para poder utilizar as bases de dados train.csv e test.csv nas análises e comparações que serão feitas é preciso tratar os dados nulos que encontramos.

Em nosso processo de análise de dados, não modificamos a base de dados do cliente, mas sim tratamos os dados para garantir análises eficazes. Identificamos e lidamos com valores vazios sem alterar os dados originais, utilizando técnicas estatísticas para preencher lacunas ou preenchendo-as com *Dummy Values*.

Esses tratamentos são essenciais para assegurar a integridade das análises e preparar os dados para a construção de modelos de previsão mais precisos, permitindo que forneçamos ao cliente insights confiáveis para embasar suas decisões estratégicas.

Como os dados serão usados tanto para análises gráficas quanto para análises preditivas optamos por realizar dois tipos de tratamentos diferentes, uma base de dados que vai incluir dummy values e uma que utilizará a mediana para a idade. Mais a frente explicaremos o porque dessa decisão.

 # **CÁLCULO DE SOBREVIVÊNCIA**

````python
#CRIAÇÃO DE NOVO DATAFRAME PARA GUARDAR OS VALORES DA PREVISÃO
df3 = pd.DataFrame()
# FILTRANDO OS VALORES DO DATAFRAME
df3['Feminino'] = titanic_previsao_mediana['Sex'].apply(lambda x: 1 if x == "female" else 0)
df3['2 Classe'] = titanic_previsao_mediana['Pclass'].apply(lambda x: 1 if x == 2 else 0)
df3['1 Classe'] = titanic_previsao_mediana['Pclass'].apply(lambda x: 1 if x == 1 else 0)
#RESULTADO DA APLICAÇÃO DA FÓRMULA
df3['res'] = -1.33 + (2.55 * df3['Feminino']) + (1.27 * df3['2 Classe']) + (2.58 * df3['1 Classe']) - (0.04 * titanic_previsao_mediana['Age'])
````

 # **PROBABILIDADE DE SOBREVIVÊNCIA**



````python
df3['Prob Sobrevivencia'] = (2.71**df3['res']) / (1 + 2.71**df3['res'])

df3.head()
````

````python
# Verificação se o passageiro tem probabilidade acima de 0.5, caso sim, ele coloca na coluna sobreviventes.
lista3 = []
for i in df3['Prob Sobrevivencia']:
  if i > 0.5:
    lista3.append(1)
  else:
    lista3.append(0)

df3['Sobreviveu'] = lista3
df3.head()
````

````python
#Inserindo a coluna sobrevintes.
titanic_previsao_mediana['Sobreviveu'] = lista3
titanic_previsao_dummy['Survived'] = lista3
````

 # **GRÁFICO 6 - BASE TITANIC_TREINO**
O número de passageiros e percentual dos que sobreviveram ou não.

````python
#Calcular o número total de passageiros
total_passengers = titanic_treino_dummy.shape[0]

#Calcular o número de passageiros que sobreviveram e não sobreviveram
survived_passengers = titanic_treino_dummy['Survived'].value_counts()[1]
not_survived_passengers = titanic_treino_dummy['Survived'].value_counts()[0]

#Calcular a porcentagem de passageiros que sobreviveram e não sobreviveram
percent_survived = (survived_passengers / total_passengers) * 100
percent_not_survived = (not_survived_passengers / total_passengers) * 100

#Definir as cores
cores = ['#029daf', '#e32551']

#Criar o gráfico de barras com as cores específicas
ax = sns.barplot(x=['Não sobreviveu', 'Sobreviveu'], y=[not_survived_passengers, survived_passengers], hue=['0', '1'], palette=cores, dodge=False)

#Adicionar título e rótulos aos eixos
plt.title('Número de passageiros que sobreviveram')
plt.ylabel('Número de Passageiros')
plt.text(x=0, y=250, s=f'{percent_not_survived:.1f}%')
plt.text(x=1, y=150, s=f'{percent_survived:.1f}%')
for i in ax.containers:
    ax.bar_label(i,)
#Exibir o gráfico
plt.show()
````
